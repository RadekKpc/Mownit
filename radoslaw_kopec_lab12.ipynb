Zaimplementuj metodę Rungego-Kutty czwartego rzędu (rozdział 10.2, Kincaid i Cheney), a następnie:<br>
* Opisz zalety metody Rungego-Kutty w porównaniu do metody z szeregami Taylora.
* Rozwiąż zagadnienie początkowe dane równaniem x' = x/t + tsec(x/t) z warunkiem początkowym x(0) = 0. Przedłuż rozwiązanie do t = 1 z krokiem h = 2^−7. Porównaj wynik z dokładnym rozwiązaniem: x(t) = t ∗ arcsin(t).

* Używając tej samej metody rozwiąż zagadnienie początkowe dane równaniem x' = 100(sin(t) − x) z warunkiem początkowym x(0) = 0 na przedziale [0, 3] używając kroków o rozmiarach h = 0.015, 0.02, 0.025, 0.03. Opisz z czego wynikają różnice w rozwiązaniach.

Definicja funkcji rungego-kutty:

```python
def runge_kutta_order_4(f, t, x, h, n):

    ta = t
    for j in range(1,n + 1):
        if t != 0:
            k1 = h*f(t,x)
            k2 = h*f(t + 0.5*h, x + 0.5*k1)
            k3 = h*f(t + 0.5*h, x + 0.5*k2)
            k4 = h*f(t + h, x + k3)

            x = x + (1.0 / 6.0 )*(k1 + 2*k2 + 2* k3 + k4)
        t = t + h
    return t,x
```

Prawidłowe rozwiązanie funkcji:

```python
x = []
y = []
for i in np.arange(-1,1,0.001):
    x.append(i)
    y.append(i*math.asin(i))
plt.plot(x,y,'bo',markersize=1)
plt.show()
```

Nasze rozwiązani, używające metody rungego-kutty:

```python
x = []
y = []
t = 0
x0 = 0
for j in np.arange(-1,1,0.01):
    n = (int) (math.fabs((j-t)/(math.pow(2,-7))))
    dz = (math.fabs((j-t)/(math.pow(2,-7))))
    h = (j-t)/dz
    a, b = runge_kutta_order_4( (lambda t,x: x/t + t*(1/math.cos(x/t))) ,t ,x0 ,h ,n)
    x.append(a)
    y.append(b)

plt.plot(x,y,'ro',markersize=1)
plt.show()
```

Dla kroków o różnych rozmiarach:

```python
for h in [0.015, 0.02, 0.025, 0.03]:
    x = []
    y = []
    t = 0
    x0 = 0
    for j in np.arange(0,3,0.001):
        n = (int) (math.fabs((j-t)/h))
        dz = (math.fabs((j-t)/h))
        if dz != 0:
            h = (j-t)/dz
            a, b = runge_kutta_order_4( (lambda t,x: 100*(math.sin(t) - x)) ,t ,x0 ,h ,n)
            x.append(a)
            y.append(b)

    print("Dla h = ",h)
    plt.plot(x,y,'go',markersize=1)
    plt.show()
```

Jak widać dla h = 0.3 wynik się psuje. Różnice w rozwiązaniach biorą się z tąd, że gdy bierzemy większy krok to otrzymujemy mniejszą dokładność, t zwiększamy zawzse o h co wiąże się z mniejszą liczbą iteracji, a co za tym idzie mniej dokłądnym rozwiązaniem.

<b>Zadanie 2 Adaptacyjna metoda Rungego-Kutty-Fehlberga</b> <br><hr>

Definicja Funkcji pomocnicznej signum:

```python
def signum(h):
    if h > 0:
        return 1
    if h < 0:
        return -1

    return 0
```

Definiujemy metode Runge–Kutta–Fehlberg :

```python
def runge_kutta_45(f,t,x,h):
    k1 = h*f(t,x)
    k2 = h*f(t + 0.25*h,x + 0.25*k1)
    k3 = h*f(t + 0.375*h,x + 0.09375*k1 + 0.28125*k2)
    k4 = h*f(t + 12/13*h,x + 1932/2197*k1 - 7200/2197*k2 + 7296/2197*k3)
    k5 = h*f(t + h, x + 439/216*k1 - 8*k2 + 3680/513*k3 - 845/4104*k4)
    k6 = h*f(t + 0.5*h, x + -8/27*k1 + 2*k2 - 3544/2565*k3 + 1859/4104*k4 - 0.275*k5)

    x4 = x + 25/216*k1 + 1408/2565*k3 + 2197/4104*k4 - 0.2*k5
    x = x + 16/135*k1 + 6656/12825*k3 + 28561/56430*k4 - 0.18*k5 + 2/55*k6
    t = t + h
    e = math.fabs(x - x4)
    return x,t,e
```

Testujemy ją dla przykłądu z książki "Numerical Mathematics and Computing 6th edition" Cheney & Kincaid 10.3

```python
a = 1.0
b = 1.5625
h = (b -a)/1000
x = 2.0
t = a
f = lambda t,x : 2.0 + (x - t - 1.0)**2
for k in range(1,1000):
    x,t,e = runge_kutta_45(f,t,x,h)
print(x,t,e)
```

Definiujemy adaptacyjną metode rungego-kutty

```python

def adaptive_runge_kutta(f,t,x,h,tb,itmax,emax,emin,hmin,hmax):
    delta = 0.5*0.00001
    iflag = 1
    k = 0
    while k <= itmax:
        k = k + 1
        if math.fabs(h) < hmin:
            h = hmin* signum(h)

        if math.fabs(h) > hmax:
            h = hmax* signum(h)

        d = math.fabs(tb - t)

        if d <= math.fabs(h):
            iflag = 0
            if d <= delta* max(math.fabs(tb),math.fabs(t)):
                break
            h = signum(h)*d

        xsave = x
        tsave = t

        x,t,e = runge_kutta_45(f,t,x,h)
        if iflag == 0:
            break
        if e < emin:
            h = 2*h
        if e > emax:
            h = h/2
            x = xsave
            t = tsave
            k = k -1

    return t,x
```

Testujemy ją dla przykładu z książki "Numerical Mathematics and Computing 6th edition" Cheney & Kincaid 10.3

```python
a = 0.0
tb = 10.0
x = 0.0
h = 0.01
t = 0.0
itmax = 1000
emax = 0.00001
emin = 0.00000001
hmax = 1.0
hmin = 0.000001
f = lambda t,x: 3 + 5*math.sin(t) + 0.2*x
print(adaptive_runge_kutta(f,a,x,h,tb,itmax,emax,emin,hmin,hmax))
```
