Zdefiniowanie Przedziału

Zdefiniowanie Frequency (Monobit) Test dla ciągu bitów: <br> Funkcje Pomocnicze:


* funkcja zamieniająca floata na ciąg bitów:

```python
def binary(num):
    return ''.join(bin(ord(c)).replace('0b', '').rjust(8, '0') for c in str(struct.pack('!f', num)))
```

* Funkcja licząca P-value:

```python
def frequency(bits):
    s = 0
    for l in bits:
        if l == '0':
            s -= 1
        if l == '1':
            s += 1
    sobs = s/(len(bits))
    res = (erfc(math.fabs(sobs)/math.sqrt(2)))
    print(res)
    if res >= 0.1 :
        print("Accept the sequence as random")
    else:
        print("Not accept the sequence as random")

    return res
```

* Funckja testująca

```python
def test_generators(N):
    twister = []
    pcg64 = []
    n_bins = 10
    
    # LOSOWANIE LICZB
    
    for _ in range(N):
        twister.append(random())
        rng = np.random.default_rng()
        pcg64.append(rng.random())
    
    # WIZUALIZACJA
    
    fig , axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
    axs[0].hist(twister, bins=n_bins, color  = 'b', label = "twister")
    axs[0].legend()
    axs[1].hist(pcg64, bins=n_bins, color  = 'r', label = "pcg64")
    axs[1].legend()
    plt.show()
    
    # SPRAWDZANIE NIEROWNOSCI Xi < Xi +1 
    
    twister_count = 0
    pcg64_count = 0
    for i in range(N-1):
        if twister[i] < twister[i+1]:
            twister_count += 1
        if pcg64[i] < pcg64[i + 1]:
            pcg64_count += 1

    print(f'N = {N}')
    print(f'dla pcg64 rownanie spełnia {pcg64_count} liczb\n')
    print(f'dla twister rownanie spełnia {twister_count} liczb\n')
    
    #Frequency (Monobit) Test

    twister_bit_string = ""
    pcg64_bit_string = ""
    for i in pcg64:
        pcg64_bit_string += binary(i)

    for i in twister:
        twister_bit_string += binary(i)

    frequency(pcg64_bit_string)
    frequency(twister_bit_string)
    
```

Testowanie generatorów:

```python
for n in N:
    test_generators(n)
```

<b>Zadanie 2 Generowanie liczb z rozkładu normalnego</b> <br><hr>
Zaimplementuj generator liczb losowych pochodzących z rozkładu normalnego za pomocą metody Boksa-Mullera i zaprezentuj wyniki dla n = 10, 100, 5000 wygenerowanych liczb. Funkcja powinna dodatkowo przyjmować wartość oczekiwaną oraz odchylenie standardowe jako parametry.
* Zwizualizuj na wykresie rozkład liczb w 10 równych przedziałach. Na tym samym wykresie umieść funkcję gęstości prawdopodobieństwa rozkładu normalnego o takich samych parametrach.
*  Wykonaj test statystyczny Shapiro-Wilka, przedstaw uzyskaną p-wartość, informację czy hipoteza zerowa została przyjęta/odrzucona dla α = 0, 05 i czy w związku z tym są podstawy do stwierdzenia że liczby pochodzą z rozkładu normalnego.


Generowanie rozkłądu normalnego N(0,1):

```python
NX = []
NY = []
for i in range(-10000,10000):
    NX.append(4*i/10000)
    NY.append((1/math.sqrt(2*math.pi))*math.exp(-math.pow(4*i/10000,2)/2))

plt.plot(NX,NY,'bo',markersize=1)
plt.show()
```

Funkcja realizująca treść zadania:

```python
def gen_norm(n,u,e):

#   METODA BOKSA-MULLERA
    
    X = []
    Y = []
    for i in range(n):

        while True:
            X1 = random()
            X2 = random()

            V1 = (float) (2*X1 - 1)
            V2 = (float) (2*X2 - 1)

            R = math.sqrt(math.fabs(V1*V1 - V2*V2))
            if R  < 1:
                break

        X.append(u + e*math.sqrt((-2)*math.log(X1))*(V1/R))
        Y.append(u + e*math.sqrt((-2)*math.log(X1))*(V2/R))
    
#     WIZUALIZACJA

    _ , axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
    axs[0].hist(X, bins=10, color  = 'b', label = "twister")
    axs[0].plot(NX,np.array(NY)*n,'ro',markersize=1)
    axs[0].legend()
    axs[1].hist(Y, bins=10, color  = 'r', label = "pcg64")
    axs[1].plot(NX,np.array(NY)*n,'bo',markersize=1)
    axs[1].legend()
    plt.show()

    
#   TESTY SHAPIRO-WILKA
    
    print(f"Wynik testu shapiro-wilka dla N = {n}:")
    WX = stats.shapiro(X)
    print(WX)
    WY = stats.shapiro(Y)
    print(WY)
```

```python
for n in N:
    gen_norm(n,0,1)
```

<b>Wnioski:</b>

* Wartośći zwracane przez test shapiro-wilka potwiedzaja skuteczność metody BOKSA-MULLERA, podna wartośći ciągi liczb należą do rozkłądu normalnego. Wartość statystyki W na poziomie istotności 0,05 i odpowiednich n jest mniejsza niż otrzymana z tablicy wartosci krytycznej nie mamy powodu odrzucić hipotezy, że próba pochodzi z rozkładu normalnego.

<b>Zadanie 3 Metoda Monte Carlo</b> <br><hr>
Napisz funkcję która zwróci przybliżoną wartość liczby pi wykorzystującą metodę Monte
Carlo do obliczenia pola powierzchni koła jednostkowego. Funkcja powinna przyjmować
liczbę wylosowanych punktów jako parametr. Zwizualizuj koło, zaznaczając na jednym
kolorem punkty które trafiły do jego wnętrza, a drugim te które trafiły na zewnątrz. Na
wykresie zwizualizuj błąd bezwzględny wraz z rosnącą liczbą punktów


```python
def monte_carlo(n,visualize = True):

    in_circle_x = []
    in_circle_y = []

    out_circle_x = []
    out_circle_y = []

    for i in range(n):
        x = np.random.uniform(-1,1)
        y = np.random.uniform(-1,1)

        if x*x + y*y <= 1:
            in_circle_x.append(x)
            in_circle_y.append(y)
        else:
            out_circle_x.append(x)
            out_circle_y.append(y)

    if visualize:
        plt.plot(in_circle_x,in_circle_y,'ro',markersize=1)
        plt.plot(out_circle_x,out_circle_y,'bo',markersize=1)
        plt.show()

    return (4*len(in_circle_x)/n)
```

Generowanie wartosci liczby PI dla różnych rzędów N:

```python
print(monte_carlo(10))
print(monte_carlo(100))
print(monte_carlo(1000))
print(monte_carlo(10000))
print(monte_carlo(100000))
print(monte_carlo(1000000))
```

* Widzimy, że punkty jednolicie rozkładają się na całym pobszarze co pozwala obliczyć pole koła z następnie liczbę PI

Wykres liczby losowań od błędu bezwzględnego: (Uwaga Liczy sie jakies 30s)

```python
X = []
Y = []
for i in range(100,10000,10):
    Y.append(math.fabs(monte_carlo(i,visualize=False)-math.pi))
    X.append(i)

plt.plot(X,Y,'bo',markersize=1)

plt.show()
```

<b>Wnioski:</b>

* Błąd bezwzględny obliczeń maleje wraz ze wzrostem liczby losowań, chociaż w dalszym ciagu jest rozrzucony, największy jest dla barzdo małej liczby losowań.  Zależy on napenwo od jakośći generatora.
